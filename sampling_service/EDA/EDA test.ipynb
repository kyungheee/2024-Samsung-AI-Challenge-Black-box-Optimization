## 2. 데이터 개별의 속성값 관찰

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create histograms for each variable with more bins
fig, axes = plt.subplots(4, 3, figsize=(20, 20))
axes = axes.ravel()  # Converts a two-dimensional array to a one-dimensional array

for i, col in enumerate(train.columns):
    sns.histplot(train[col], ax=axes[i], kde=True, bins=40)  # Adjust the number of bins here
    axes[i].set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()

# Create box-plots for each variable
fig, axes = plt.subplots(4, 3, figsize=(20, 20))
axes = axes.ravel()

for i, col in enumerate(train.columns):
    sns.boxplot(y=train[col], ax=axes[i])
    axes[i].set_title(f'Box-plot of {col}')
    
plt.tight_layout()
plt.show()
```

## 3. 속성 간의 관계 분석 

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming `train` is your DataFrame

# Calculate the correlation matrix for available features
corr_matrix = train.corr()
print('Feature Correlation Matrix')
print(corr_matrix.sort_values(ascending=True))

# Visualize the correlation matrix
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Pairplot to visualize scatter plots between features
sns.pairplot(train, diag_kind='kde', kind='scatter')
plt.show()
```
